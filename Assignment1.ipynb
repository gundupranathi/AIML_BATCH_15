## Data visualization using matploilib


```python
people=['Ann','Brandon','David','julio','lily']
age=[21,12,32,45,37,18,28]
weight=[55,25,35,77,68,15]
height=[160,135,170,165,173,168]
import matplotlib.pyplot as plt

plt.scatter(weight,height)
plt.title("Relationship between Weight and Height of Patients")
plt.ylabel("Height in Meters")
plt.xlabel("Weight of Patients in Kgs")
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.title("People weight in kgs",fontsize=16,fontstyle='italic')
people=['Jahnavi','Harshini','Pinky','Rinky','Sweety']
height=[170,180,160,175,185]
weight=[70,80,40,30,20]
plt.bar(x=people, height=weight, width=0.6,color='orange',edgecolor='k',alpha=0.6)
plt.xlabel("People",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("Weight(in kgs)",fontsize=15)
plt.show()
```

```python
import numpy as np
plt.figure(figsize=(7,5))
plt.hist(weight,color='red',edgecolor='k',alpha=0.75,bins=5)
plt.title("Histogram of patient weight",fontsize=18)
plt.xlabel("Weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
